```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
df = pd.read_csv("data/survey_results_public_2022.csv", index_col="ResponseId")
schema_df = pd.read_csv("data/survey_results_schema.csv", index_col="qname")
```

```python
df.head()
```

```python
# Rename the ConvertedCompYearly column label to SalaryUSD
# Note that its good practice to check the df before applying changes
df_temp = df.rename(columns={"ConvertedCompYearly": "SalaryUSD"})
display(df_temp.head())

# After confirming, apply the change.
df.rename(columns={"ConvertedCompYearly": "SalaryUSD"}, inplace=True)
display(df.head())
```

```python
# Check how many non-null salaries in df
df["SalaryUSD"].count()
```

```python
# Replace Yes No answers to booleans in the "TBranch" column
# TBranch answers the question: Would you like to participate in the Professional Developer Series?
# According to the schema.
df_temp = df["TBranch"].replace({"Yes": True, "No": False})

# Apply change
df["TBranch"].replace({"Yes": True, "No": False}, inplace=True)
display(df["TBranch"])

# Find out how many wants to join the Professional Developer Series
n_true = df.loc[df["TBranch"] == True]["TBranch"].count()
display(n_true)
```
